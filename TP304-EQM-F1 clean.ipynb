```python
#Importar Librarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import datetime as dt
import seaborn as sns
```

## Carga y limpieza de datos

```python
#open files 2018
df2018=pd.read_csv("delitos_2018.csv")
df2018.head()
```

```python
#open files 2019
df2019=pd.read_csv("delitos_2019.csv")
df2019.info()
```

```python
#open files 2020
df2020=pd.read_csv("delitos_2020.csv")
df2020.info()
```

```python
#open files 2021
df2021= pd.read_csv("delitos_2021.csv")
df2021.info()
```

```python
#open files 2022
df2022=pd.read_csv("delitos_2022.csv")
df2022.info()
```

## Unificar data

```python
df_CABA= pd.concat([df2018,df2019,df2020,df2021,df2022],axis=0)
df_CABA.to_csv("df_CABA.csv", index=False)
df_CABA
```

```python
df_CABA.info()
```

```python
#Verificar duplicados
df_CABA.duplicated().sum(axis=0)
```

```python
#Verificar nulos
df_CABA.isnull().sum(axis=0)
```

```python
#Convertir numero de Comuna,Franja a entero
df_CABA.comuna= df_CABA.comuna.astype('Int64')
df_CABA.franja= df_CABA.franja.astype('Int64')
```

```python
#Convertir fechas
df_CABA['fecha'] = pd.to_datetime(df_CABA['fecha'], format='%Y-%m-%d')
print(df_CABA['fecha'])
```

```python
#Convertir tipo de fecha anio, mes
df_CABA['anio'] = df_CABA['fecha'].dt.year
df_CABA['mes'] = df_CABA['fecha'].dt.month
df_CABA.info()
#pendiente convertir  dia, franja
```

```python
#Verificar si son los mismos datos faltantes en columnas Latitud y Longitud 
df_CABA[(df_CABA['latitud'].isna() & df_CABA['longitud'].isna())]
```

```python
#Eliminar datos nulos y columnas 

df_CABA_clean= df_CABA.dropna(subset= ['latitud','longitud','comuna','franja'])

df_CABA_clean= df_CABA_clean.drop(columns=['id-mapa','cantidad'])
```
